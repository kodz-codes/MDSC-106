```python
# B. Make sure all the variables with continuous values are of ‘Float’ type. [For Example: MonthlyCharges, TotalCharges]
# checking the data types of the df.
new_df.info()
```

```python
# from the above data information we can understand that only MonthlyCharges are of float datatype, And TotalCharges is of Object type. We will try to vonvert it into float data type.
new_df['TotalCharges'] = pd.to_numeric(new_df['TotalCharges'],errors='coerce')
new_df.info()
```

```python
# Plotting pie-charts for each feature in the dataframe using .pie() function.
import matplotlib.pyplot as plt

def pie_charts(data_frame):
  continuous_cols = data_frame.select_dtypes(include=['float', 'int']).columns
  counts = []
  i , j = 0 , 0

  fig , ax = plt.subplots(4,4,figsize=(15,10))
  fig.subplots_adjust(hspace=0.5)

  for f in data_frame.columns:
    if f not in continuous_cols and f != 'customerID':
      ax[i,j].pie(x=data_frame.groupby(f).size(),labels=data_frame[f].unique().tolist(),autopct='%1.1f%%',radius=1.2)
      ax[i,j].set_title(f)
      j = j + 1

    if j > 3:
      j = 0
      i = i + 1

pie_charts(new_df)
```
