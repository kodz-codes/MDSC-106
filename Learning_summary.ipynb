```python
# Adding random values to the new columns added to the dataframe
import random

# Random values list
names_list = ['Abhinay','Jaswath','Swaroop','Ayush','Chetri','Sumiran','Pankaj','Bhargav']

# Filling the random values of random list into the new column added.
for i in range(0,442):
  temp_variable = random.choice(names_list)
  temp_aadhaar = random.randint(10000000000,100000000000)
  # print(variable)
  df.loc[i,'Name_of_patient'], df.loc[i,'Patient_Aadhaar_number'] = temp_variable, temp_aadhaar
  # print(df['Name_of_patient'])

df
```

```python
# We can handle the missing values using .isnull() and .fillna() functions of the dataframe
df.isnull()  # displays True wherever the values are null.
df.fillna(23) # fills all null values with value 23 in the entire dataframe
```

```python
# We can remove the duplicate values in the Dataframe by using .drop_duplicates(inplace = True) function.
df.drop_duplicates(inplace = True)
df
```

```python
# We can filter data using dataframes as follows. Diaplying the filtered data
filtered_data = df[df['age'] > 30]
filtered_data
```

# **Line Plot**

```python
# Drawing a lineplot between age and their coresponding bp
sns.lineplot(data = df, x = df['age'], y = df['bp'])
# So from seeing the line plot we can clearly understand that people under the age of 60-70 are having high blood pressure.
```

# **Satter Plot**

```python
# Plotting a scatter plot betweeen age and bmi.
sns.scatterplot(data = df, x = df['age'], y = df['bmi'])
# we can understand that people under the category of 30-50 are having higher Body-Mass-Index compared to other.
```

# **Bar plots**

```python
# Plotting a bar plot for the attributes, sex and bmi.
x = df['sex']
y = df['bmi']
# plt.bar(df['age'], df['target'])
plt.bar(x,y)
plt.show()
# We can understand that one category of sex is having greater bmi compared to others., We are doing because to compare which sex is having greater bmi.
```

# **Stacked Bar Plot**

```python
# We will try to plot the stacked bar plot between two sexs.
# plot bars in stack manner
# plot bars in stack manner
x = df['sex']
y1 = df['age']
y2 = df['bmi']
y3 = df['bp']

plt.bar(x, y1, color='yellow')
plt.bar(x, y2, bottom=y1, color='violet')
plt.bar(x, y3, bottom=y1+y2, color='skyblue')
plt.xlabel("Sex")
plt.ylabel("Their overall body health")
plt.legend(["Age","BMI","BP"])
plt.title("Average health of a particular sex")
plt.show()
```

# **Box Plot**

```python
# We will try to plot a Box Plot between Sex and s3 attribute, Sex and age attribute.
plt.figure(figsize = (15,5))
plt.subplot(1,3,1)
sns.boxplot(data = df, x = df['sex'], y = df['s3'])

plt.subplot(1,3,2)
sns.boxplot(data = df, x = df['sex'], y = df['age'])
```

# **Histogram Plot**

```python
# We will try to plot a histogram plot for the age of the people.
sns.histplot(data = df, x = df['age'])
# The below histplot shows the count of people under a particular age
```

```python
plt.figure(figsize=(20, 7)) # The figure() sttribute of matplotlib can be used when we want to change the default size of the plot.
# plt.show()

plt.subplot(1,4,1) # This help us to draw a subplot in a plot indicating the one row, four columns, and the index of the subplot respectively
#1st row, with 4 columns and 1st subplot
sns.histplot(df['age'], kde = True,color='r')
plt.title('Histogram')

# scatterplot
plt.subplot(1, 4, 2)
plt.scatter(df['age'],df['bp'],color = 'g')
plt.title('Scatterplot')

# boxplot
plt.subplot(1, 4, 3)
sns.boxplot(y=df['target'],color = 'b')
plt.title('Boxplot')

# barplot
plt.subplot(1, 4, 4)
sns.barplot(x='target', y = 'bmi', data = df)
plt.title('Barplot')

plt.show()
```
